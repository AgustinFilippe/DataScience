El <b>Análisis Exploratorio de Datos (EDA)</b> es un enfoque o técnica estadística para analizar conjuntos de datos con el fin de resumir sus características importantes y principales, generalmente mediante el uso de algunas ayudas visuales. El enfoque AED se puede utilizar para obtener conocimiento sobre los siguientes aspectos de los datos:
    - Características principales o rasgos de los datos.
    - Las variables y sus relaciones.
    - Descubrir las variables importantes que se pueden utilizar en nuestro problema.

AED es un enfoque iterativo que incluye:

    - Generar preguntas sobre nuestros datos.
    - Buscar las respuestas mediante la visualización, transformación y modelado de nuestros datos.
    - Utilizar las lecciones que aprendemos para refinar nuestro conjunto de preguntas o generar un nuevo conjunto de preguntas.

Análisis Exploratorio de Datos en R

En el lenguaje R, vamos a realizar AED bajo dos clasificaciones amplias:

    - Estadísticas descriptivas, que incluyen la media, mediana, moda, rango intercuartílico, entre otros.
    - Métodos gráficos, que incluyen histogramas, estimación de densidad, diagramas de caja, entre otros.


<b>¿Por qué análisis de datos exploratorios (EDA)?</b>

    EDA es un paso de análisis de datos para comprender mejor los datos sin suposiciones. Durante EDA, busca patrones de datos, tendencias y resume las características principales del conjunto de datos.

    EDA comienza haciendo preguntas relacionadas con partes específicas de los datos para realizar una investigación de datos como un detective de datos.
    Como ejemplo, tenemos los datos de arribo y salida de vuelos junto con datos meteorológicos. Se puede preguntar cómo el clima afecta la puntualidad de las llegadas/salidas del vuelo, qué temperaturas y condiciones de viento son ideales para volar en invierno y las condiciones que retrasan el vuelo.
    Durante el proceso de EDA, recopile tantas preguntas que inviten a la reflexión como sea posible relacionadas con los datos sin ningún juicio; nunca se sabe qué conocimientos le esperan para ser explorados en el conjunto de datos.

<b>Objetivo de análisis de datos exploratorios</b>

    - Obtener una visión detallada del conjunto de datos.
    - Descubrir patrones, relaciones y tendencias de datos subyacentes
    - Identificar variables críticas en el conjunto de datos.
    - Detectar valores atípicos
    - Identificar datos faltantes

<b>¿Cómo realizar EDA?</b>
El análisis de datos exploratorios tiene tres pasos básicos.

    - Observar los valores de los datos sin procesar.
    - Calcular estadísticas resumidas como la media, la mediana, la desviación estándar y el rango intercuartil.
    - Crear visualizaciones de datos trazando diferentes tipos de gráficos para responder preguntas interesantes sobre los datos.

Calcular estadísticas resumidas del conjunto de datos.
Las estadísticas de resumen son valores numéricos únicos que resumen una gran cantidad de valores.

Aquí usamos el operador de pipeline %>% que combina múltiples operaciones en una única cadena secuencial de acciones y hace que el código sea legible y comprensible.

%>% toma la salida de una función y luego la canaliza para que sea el argumento de entrada de la siguiente función; Como en el siguiente ejemplo, pasamos el marco de datos data_flights a la función summarize().
La función summarize() toma muchos valores y los resume en un solo valor.


```r
distance_summary <- data_flight %>% 
  summarize(
    min = min(DISTANCE, na.rm = TRUE),
    q1 = quantile(DISTANCE, 0.25, na.rm = TRUE),
    median = quantile(DISTANCE, 0.5, na.rm = TRUE),
    q3 = quantile(DISTANCE, 0.75, na.rm = TRUE),
    max = max(DISTANCE, na.rm = TRUE),
    mean = mean(DISTANCE, na.rm = TRUE),
    sd = sd(DISTANCE, na.rm = TRUE),
    missing = sum(is.na(DISTANCE))
  )
View(distance_summary)
```

En el código anterior, tomamos todos los valores de la columna DISTANCE y luego generamos las estadísticas de resumen, La media también se llama promedio para la mediana de la columna, el valor medio, el valor mínimo y máximo en la columna, los valores del primer y tercer cuartil, Desviación estándar y Cualquier valor faltante en la columna. Los mismos resultados también se pueden generar sin usar un operador de tubería, como se muestra a continuación.

```r
distance_summary <-summarize(data_flight,
  min = min(DISTANCE, na.rm = TRUE),
  q1 = quantile(DISTANCE, 0.25, na.rm = TRUE),
  median = quantile(DISTANCE, 0.5, na.rm = TRUE),
  q3 = quantile(DISTANCE, 0.75, na.rm = TRUE),
  max = max(DISTANCE, na.rm = TRUE),
  mean = mean(DISTANCE, na.rm = TRUE),
  sd = sd(DISTANCE, na.rm = TRUE),
  missing = sum(is.na(DISTANCE))
)
```

skim() es otra función que proporciona estadísticas resumidas del marco de datos especificado. Devuelve un objeto skim_df que enumera cada tipo de datos por separado.

Ahora, seleccionemos solo algunas columnas de interés para estudiar y filtrar los registros donde el ORIGIN es ORD.

```r
chicago_flights <- data_flight %>% 
  select(DAY_OF_MONTH,OP_CARRIER_FL_NUM,OP_CARRIER, ORIGIN, DEST,
         DEP_TIME,ARR_TIME, DISTANCE, DEP_DEL15, ARR_DEL15, CANCELLED  ) %>%
  filter(ORIGIN=='ORD')
View(chicago_flights)
```

select() le permite elegir un subconjunto de columnas de un marco de datos
select() permite seleccionar un subconjunto de columnas de un marco de datos
filter() filtra solo las filas que coinciden con los criterios especificados.

En el código anterior, seleccionamos algunas columnas DAY_OF_MONTH, OP_CARRIER_FL_NUM, OP_CARRIER, ORIGIN, DEST, DEP_TIME, ARR_TIME, DISTANCE, DEP_DEL15, ARR_DEL15, CANCELED de data_flights y luego canalizamos esa salida para filtrar los registros donde los vuelos se originaron desde ORD(O 'Hare Chicago) en un nuevo dataframe chicago_flights.


<b>1. ¿Cuántos vuelos de Chicago a JFK en enero 2020?</b>

Para obtener el número total de vuelos entre ORD y JFK en enero de 2020, canalizamos chicago_flights al summarize()

```r
num_flights <- chicago_flights %>% summarise(count=n())
print(num_flights)
```

<b>2. ¿Qué aerolíneas operaron entre ORD y JFK y cuántos vuelos operó cada una de ellas en enero de 2020?</b>

Calcule el número de vuelos de cada aerolínea agrupando primero por OP_CARRIER y luego resumiendo los recuentos de cada aerolínea.

```r
carriers <- chicago_flights %>% 
  group_by(OP_CARRIER) %>%
  summarise(count=n())
View(carriers)
```

Visualizando las diferentes aerolíneas y la cantidad de vuelos que tienen entre ORD y JFK usando un gráfico de barras usando geom_col()

```r
ggplot(data = carriers, mapping = aes(x = OP_CARRIER, y= count)) +
  geom_col(position = "dodge", fill="pink") +
  ggtitle("Carriers between ORD and JFK") +
              xlab("Carriers") +
              ylab("count") +
  geom_text(aes(label = count))
```

Chiste nerd sobre la importancia de etiquetar los ejes!!!



<b>3. ¿Cuántos vuelos se retrasaron en su salida de Chicago a JFK?</b>

Filtra todos los vuelos de Chicago con destino a JFK y luego agrupa por DEP_DEL15, que indica si la salida se retrasó desde el Origen 15 min o más.

```r
flights_to_jfk <- chicago_flights %>% filter(DEST=='JFK') %>%
  group_by(DEP_DEL15) %>%
  summarise(count=n()) 
View(flights_to_jfk)
```

Borrar cualquier registro con datos faltantes en el DEP_DEL15 utilizando drop_na()

```r
flights_to_jfk <-flights_to_jfk %>% drop_na(DEP_DEL15)
```

Cambiar las etiquetas para leer mejor la semántica... 0 indica que la salida fue a tiempo, 1 que la salida estuvo demorada.

```r
flights_to_jfk$DEP_DEL15[flights_to_jfk$DEP_DEL15==0]="On Time"
flights_to_jfk$DEP_DEL15[flights_to_jfk$DEP_DEL15==1]="Delayed"
View(flights_to_jfk)
```

```r
# display the times the xna flights were on time or delayed
ggplot(data= flights_to_jfk, 
       mapping= aes(x=DEP_DEL15, y=count)) +
  geom_col(fill="green") +
   ggtitle('JFK flights on time/delay')+ 
  xlab('Ontime/delay')+ 
  ylab('Count') + 
  geom_text(aes(label=count))
```

<b>4. ¿Cuántos vuelos retrasados estaban fuera del rango normal y de qué compañías aéreas?</b>

Para encontrar los vuelos retrasados, calcule la duración del vuelo tomando la diferencia entre la hora de llegada y la hora de salida.

Filtre los vuelos de Chicago cuyo destino sea JFK y luego cree un nuevo campo llamado duración usando mutate() del paquete dplyr.

mutate() le permite crear o calcular nuevas variables basadas en una variable existente.

```r
flights_to_jfk_dur <- chicago_flights %>% 
  filter(DEST=='JFK') %>%
  mutate(duration =ifelse( as.numeric(ARR_TIME) > as.numeric(DEP_TIME), 
                  as.numeric(ARR_TIME) - as.numeric(DEP_TIME),
                 (2400 + as.numeric(ARR_TIME) )- as.numeric(DEP_TIME)))
flights_to_jfk_dur <- flights_to_jfk_dur %>%
  drop_na(duration)
```

Cuando tenemos una hora de llegada después de la medianoche, obtenemos valores negativos para la duración. Por lo tanto usamos un ifelse() dentro de mutate()

También eliminamos las filas a las que les falta duración.

Un diagrama de caja es muy útil para encontrar los vuelos entre ORD y JFK que se retrasaron dentro del rango normal.

```r
ggplot(flights_to_jfk_dur, aes(y=duration)) +
  geom_boxplot() +
  ggtitle("Flight Duration Outliers ORD to JFK")
```

Este diagrama de caja nos muestra que dos vuelos se retrasaron más allá del rango normal.

Como parte de la EDA, le gustaría conocer los valores atípicos de duración de todos los diferentes transportistas que operaban entre ORD y JFK en enero de 2020.

is_outlier es una función para identificar valores atípicos. Los valores atípicos son observaciones por encima de 1,5 * IQR + tercer IQR o cualquier valor por debajo de 1,5 * IQR + primer cuartil.

```r
is_outlier <- function(x) {
  return(x < quantile(x, 0.25) - 1.5 * IQR(x) | 
           x > quantile(x, 0.75) + 1.5 * IQR(x))
}
```

Agrupar todos los vuelos entre ORD y JFK por OP_CARRIER y entonces agregar otra columna llamada "outliers" utilizando mutate() para los outliers de cada uno de las aerolíneas.

```r
flights_to_jfk_dur <- flights_to_jfk_dur %>%
  group_by(OP_CARRIER) %>%
  mutate(outlier = ifelse(is_outlier(duration), duration, as.numeric(NA)))
#Plot outliers for duration per carrier
ggplot(flights_to_jfk_dur, aes(y=duration, x= OP_CARRIER)) +
  geom_boxplot() +
  ggtitle("Outliers per Carrier for flight between ORD and JFK")
```

Para incluir los labels en los boxplots:

```r
ggplot(flights_to_jfk_dur, aes(y=duration, x= OP_CARRIER)) +
  geom_boxplot()+ 
  ggtitle("Outliers per Carrier for flight between ORD and JFK") +
  xlab("Flight Carriers") +
  geom_text(aes(label = outlier),
            na.rm=TRUE, 
            position=position_jitter(width=1,height=1))
```

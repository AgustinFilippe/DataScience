# Protocolo HTTP

### Importancia para el Web Scraping

Es esencial entender cmo funcionan las solicitudes y respuestas HTTP porque:

- **Realizar solicitudes**: Usaremos herramientas como `requests` en Python para enviar solicitudes GET y POST a los servidores y extraer los datos que necesitamos.

- **Manejo de respuestas**: Debemos interpretar correctamente el cdigo de estado y el contenido de la respuesta. Si recibimos un 200, podemos proceder a extraer datos; si es un 404, debemos manejar el error adecuadamente.

- **Encabezados y autenticacin**: Algunos sitios requieren que incluimos encabezados especficos o que autenticamos nuestra solicitud (por ejemplo, con tokens de acceso) para obtener los datos. Conocer cmo funcionan estos aspectos es crucial para acceder a informacin restringida.



HTTP (HyperText Transfer Protocol) es un protocolo de comunicacin que permite la transferencia de datos en la web. Se basa en un modelo de cliente-servidor, donde el cliente (normalmente un navegador web) solicita recursos y el servidor los proporciona.

## Estructura de una Solicitud HTTP

Una solicitud HTTP se compone de varias partes:

1. **URL (Uniform Resource Locator)**: Es la direccin del recurso que queremos acceder, junto con los parmetro. Por ejemplo, `https://www.ejemplo.com?page=1`.

2. **Mtodo**: Indica la accin que se desea realizar. Los mtodos ms comunes son:
   - **GET**: Se utiliza para solicitar datos del servidor. Por ejemplo, cuando abrimos una pgina web, el navegador enva una solicitud GET para obtener el contenido.
   - **POST**: Se usa para enviar datos al servidor, como al enviar un formulario.
<br/><br/>

3. **Encabezados (Headers)**: Proporcionan informacin adicional sobre la solicitud, como el tipo de contenido que acepta el cliente (`Accept`), informacin sobre el navegador (`User-Agent`), entre otros.

4. **Cuerpo (Body)**: Para solicitudes POST, es utilizado para enviar datos al servidor.

<br>



```python
GET /index.html HTTP/1.1
Host: www.ejemplo.com
User-Agent: Mozilla/5.0 (X11; Linux x86_64; rv:109.0)
Accept: text/html,application/xhtml+xml,application/xml
```

```python
POST /login HTTP/1.1
Host: www.example.com
User-Agent: Mozilla/5.0 (X11; Linux x86_64; rv:109.0)
Content-Type: application/x-www-form-urlencoded; charset=UTF-8
Content-Length: 28

username=john_doe&password=S3cr3tP@ss
```

## Respuesta HTTP

Cuando el servidor recibe una solicitud, enva una respuesta que tambin tiene una estructura especfica:

1. **Cdigo de estado**: Indica el resultado de la solicitud. Algunos cdigos comunes son:

2. **Encabezados de respuesta (header)**: Proporcionan informacin sobre el tipo de contenido devuelto y otros detalles.

3. **Cuerpo de la respuesta (body)**: Contiene los datos solicitados, como el HTML de una pgina web, un archivo JSON, etc.



```python
HTTP/1.1 200 OK
Content-Type: text/html
Content-Length: 1256
Set-Cookie: session_id=abc123

<!DOCTYPE html>
<html>
  <head>
    <title>Ejemplo</title>
  </head>
  <body>
    <h1>Bienvenido</h1>
    <!-- Contenido de la pgina... -->
  </body>
</html>
```

## Qu es HTTPS?

La comunicacin entre el cliente web y el servidor web est cifrada. Mediante este sistema, se evita que un tercero no autorizado tenga acceso los datos, por ejemplo, observando el trfico de la red wifi.

El servidor web se autentica enviando un certificado al navegador justo al comienzo de la transmisin de datos, que garantiza la fiabilidad del dominio. Esta medida ayuda a combatir el fraude por parte de pginas web falsas.

# Javascript

Lenguaje de programacin de alto nivel, interpretado y orientado a objetos, ampliamente utilizado para crear interactividad y dinamismo en pginas web. Desarrollado inicialmente por Netscape en 1995, JavaScript ha evolucionado para convertirse en un estndar del desarrollo web y es una de las tecnologas clave que conforman la trada fundamental del desarrollo web, junto con HTML y CSS.

### Caractersticas Principales

**Interactividad**: JavaScript permite a los desarrolladores crear aplicaciones web interactivas que responden a las acciones del usuario, como clics, desplazamientos y entradas de formularios. Esto mejora significativamente la experiencia del usuario.

**Manipulacin del DOM**: JavaScript se utiliza para manipular el Document Object Model (DOM), lo que permite modificar el contenido y la estructura de una pgina web en tiempo real sin necesidad de recargarla.

**Programacin Asncrona**: Gracias a su naturaleza no bloqueante, JavaScript puede realizar operaciones asncronas, como solicitudes a servidores, permitiendo a los desarrolladores cargar datos y actualizarlos dinmicamente en la interfaz de usuario.

# Document Object Model - DOM

El Document Object Model (DOM) es una interfaz de programacin que proporciona una representacin estructurada de documentos HTML y XML. Permite a los lenguajes de programacin, como JavaScript, interactuar con la estructura del documento, manipular su contenido y aplicar estilos dinmicamente. El DOM trata un documento como una jerarqua de nodos, donde cada elemento, atributo y texto es un objeto que puede ser accedido y modificado.
